# Understanding Code vs. Markdown Cells

This notebook demonstrates the fundamental difference between executable code and human-readable explanation. 

In Data Science, we use notebooks to tell a story:
* **Code cells** perform the heavy lifting (math, data processing, modeling).
* **Markdown cells** explain *why* we are doing it and what the results mean.

```python
# Defining some basic variables
store_name = "Predictify Retail"
daily_sales = [150, 200, 180, 220, 190]

# Calculating the total weekly sales
total_sales = sum(daily_sales)

# Displaying the output
print(f"Welcome to {store_name}!")
print(f"The total sales for the last 5 days were: {total_sales} units.")
```

### What just happened?

The cell above was a **Code cell**. 
When executed, the Python kernel calculated the sum of our `daily_sales` list and printed a string to the screen. 

Notice how this current cell (a **Markdown cell**) does not compute anything. Instead, it allows me to interpret the results for the person reading this document. If I only used `# comments` inside the code block, this notebook would be much harder to read!
